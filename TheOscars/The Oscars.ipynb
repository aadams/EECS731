```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
```

```python
AAPL = pd.read_csv("data/AAPL.csv")
AMZN = pd.read_csv("data/AMZN.csv")
FB = pd.read_csv("data/FB.csv")
GOOG = pd.read_csv("data/GOOG.csv")
```

```python
#Taking only the date, not the time
dates = AAPL.timestamp.tolist()
lst = []
for x in dates:
    lst.append(x[0:10])
AAPL['Dates'] = lst
AAPL.drop('timestamp', axis = 1, inplace = True)
AAPL.Dates = pd.to_datetime(AAPL.Dates)
AAPL = AAPL.resample('d', on='Dates').sum().dropna(how='all')
AAPL.rename({'value': 'Tweets'}, inplace = True, axis = 'columns')
```

```python
dates = AMZN.timestamp.tolist()
lst = []
for x in dates:
    lst.append(x[0:10])
AMZN['Dates'] = lst
AMZN.drop('timestamp', axis = 1, inplace = True)
AMZN.Dates = pd.to_datetime(AMZN.Dates)
AMZN = AMZN.resample('d', on='Dates').sum().dropna(how='all')
AMZN.rename({'value': 'Tweets'}, inplace = True, axis = 'columns')
```

```python
dates = FB.timestamp.tolist()
lst = []
for x in dates:
    lst.append(x[0:10])
FB['Dates'] = lst
FB.drop('timestamp', axis = 1, inplace = True)
FB.Dates = pd.to_datetime(FB.Dates)
FB = FB.resample('d', on='Dates').sum().dropna(how='all')
FB.rename({'value': 'Tweets'}, inplace = True, axis = 'columns')
```

```python
dates = GOOG.timestamp.tolist()
lst = []
for x in dates:
    lst.append(x[0:10])
GOOG['Dates'] = lst
GOOG.drop('timestamp', axis = 1, inplace = True)
GOOG.Dates = pd.to_datetime(GOOG.Dates)
GOOG = GOOG.resample('d', on='Dates').sum().dropna(how='all')
GOOG.rename({'value': 'Tweets'}, inplace = True, axis = 'columns')
```

# Do Tweets Go Up When Earnings are Reported?

```python
GOOG.plot()
plt.axvline(x = dt.date(2015, 4, 22), color = 'k', linestyle = '--')
plt.title('Earnings Date vs Amount of Tweets for Google')
plt.ylabel('Tweets')
```

# Do Tweets Correlate to Volatile Price Changes?

```python
AAPLStock = pd.read_csv("data/AAPLStock.csv")
AMZNStock = pd.read_csv("data/AMZNStock.csv")
FBStock = pd.read_csv("data/FBStock.csv")
GOOGStock = pd.read_csv("data/GOOGStock.csv")
```

```python
AAPLStock.date = pd.to_datetime(AAPLStock.date)
AMZNStock.date = pd.to_datetime(AMZNStock.date)
FBStock.date = pd.to_datetime(FBStock.date)
GOOGStock.date = pd.to_datetime(GOOGStock.date)
```

```python
#Needed Dates
#2015-02-26
#2015-04-22
```

```python
AMZNStock = AMZNStock[515:554]
FBStock = FBStock[515:552]
GOOGStock = GOOGStock[231:270]
AAPLStock = AAPLStock[515:554]
AMZNStock.set_index('date', inplace = True)
FBStock.set_index('date', inplace = True)
GOOGStock.set_index('date', inplace = True)
AAPLStock.set_index('date', inplace = True)
```

```python
ax1 = FB.plot(style = '-b', label = 'Tweets')

ax2 = ax1.twinx()
ax2.spines['right'].set_position(('axes', 1.0))
FBStock.close.plot(ax=ax2, style = '-g', label = 'Price')
plt.legend()
plt.xlabel('Date')
plt.ylabel('Price vs Tweets')
plt.title('Price vs Tweets for Facebook')
```
